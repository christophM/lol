```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
timelines = pd.read_csv("../data/timelines_df.csv")
timelines.set_index("matchId", inplace=True)
```

```python
timelines["gold_diff"] = timelines.gold_100 - timelines.gold_200
```

```python
timelines.shape
```

```python
rf = RandomForestClassifier(n_estimators=200)
```

```python
X = timelines[["gold_diff", "timestamp", "dragon_kill_100", "dragon_kill_200"]]
y = timelines["winner_100"]
```

```python
rf.fit(X, y)
```

```python
timelines["prediction"] =  [x[1] for x in rf.predict_proba(X)]
```

```python
matchId = timelines.index.unique()[54]
```

```python
match = timelines.ix[matchId]
match.plot(x="timestamp", y="prediction", ylim=(0,1))
```
