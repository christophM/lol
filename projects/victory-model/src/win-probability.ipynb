# Fit victory model

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
```

```python
timelines = pd.read_csv("../data/timelines_df.csv")
timelines.set_index("matchId", inplace=True)
```

## Compute additional features

```python
timelines["gold_diff"] = timelines.gold_100 - timelines.gold_200
timelines["dragons_100"] = timelines.groupby(level=0).dragon_kill_100.cumsum()
timelines["dragons_200"] = timelines.groupby(level=0).dragon_kill_200.cumsum()
timelines["dragons_diff"] = timelines.dragons_100 - timelines.dragons_200
timelines["gold_diff_rel"] = timelines.gold_diff / timelines.gold_100
timelines["gold_100_rel"] = timelines.gold_100 / timelines.timestamp
timelines["gold_200_rel"] = timelines.gold_200 / timelines.timestamp
timelines["level_sum_diff"] = timelines.level_sum_100 - timelines.level_sum_200
timelines["level_max_diff"] = timelines.level_max_100 - timelines.level_max_200
```

## Create training and test sets

```python
## randomly draw training and test data
np.random.seed(42)
sample_size = int(0.5 * len(timelines.index.unique()))
train_index = np.random.choice(timelines.index.unique(), size=sample_size, replace=False)
test_index = np.setdiff1d(timelines.index.unique(), train_index)
```

```python

## TODO: check why there are two missing entries for dragons_100/200/diff
## remove this line
timelines = timelines.dropna(axis = 0)
```

```python
X = timelines[["gold_diff",  "gold_100_rel","gold_100", "gold_200", "gold_200_rel", "gold_diff_rel", 
               "dragons_100", "dragons_200",  "dragons_diff", 
               "level_max_100", "level_max_200", "level_sum_100", "level_sum_200", "level_sum_diff", "level_max_diff"]]
X_train = X.ix[train_index]
X_test = X.ix[test_index]
y = timelines["winner_100"]
y_train = y.ix[train_index]
y_test = y.ix[test_index]
```

## Fit random forest

```python
rf = RandomForestClassifier(n_estimators=200, verbose=0)
rf.fit(X_train, y_train)
```

## Evaluate forest

```python
## Feature importance
zip(rf.feature_importances_, X.columns)
```

## Plot single match

```python
plot_data = timelines.ix[train_index]
plot_data["prediction"] =  [x[1] for x in rf.predict_proba(X_train)]
```

```python
matchId = plot_data.index.unique()[56]
match = plot_data.ix[matchId]
match.plot(x="timestamp", y="prediction", ylim=(0,1))
```

## Evaluate performance

```python
from sklearn.metrics import roc_auc_score
y_true = y_test
y_scores = [x[1] for x in rf.predict_proba(X_test)]
roc_auc_score(y_true, y_scores)
```
