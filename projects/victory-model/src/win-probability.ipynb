```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
```

## Load data

```python
timelines = pd.read_csv("../data/timelines_df.csv")
timelines.set_index("matchId", inplace=True)
timelines.fillna(0, inplace=True)
```

```python
timelines.describe()
```

## Compute additional features

```python
timelines["gold_diff"] = timelines.gold_100 - timelines.gold_200
timelines["dragons_100"] = timelines.groupby(level=0).dragon_kill_100.cumsum()
timelines["dragons_200"] = timelines.groupby(level=0).dragon_kill_200.cumsum()
timelines["dragons_diff"] = timelines.dragons_100 - timelines.dragons_200
timelines["gold_diff_rel"] = timelines.gold_100 / timelines.gold_200
timelines["gold_100_rel"] = timelines.gold_100 / timelines.timestamp
timelines["gold_200_rel"] = timelines.gold_200 / timelines.timestamp
timelines["level_sum_diff"] = timelines.level_sum_100 - timelines.level_sum_200
timelines["level_max_diff"] = timelines.level_max_100 - timelines.level_max_200
timelines["champion_kill_100_sum"] = timelines.groupby(level=0).champion_kill_100.cumsum()
timelines["champion_kill_200_sum"] = timelines.groupby(level=0).champion_kill_200.cumsum()
timelines["champion_kill_sum_diff"] = timelines.champion_kill_100_sum - timelines.champion_kill_200_sum
timelines["building_kill_100_sum"] = timelines.groupby(level=0).building_kill_100.cumsum()
timelines["building_kill_200_sum"] = timelines.groupby(level=0).building_kill_200.cumsum()
timelines["inhibitor_kill_100_sum"] = timelines.groupby(level=0).kill_inhibitor_100.cumsum()
timelines["inhibitor_kill_200_sum"] = timelines.groupby(level=0).kill_inhibitor_200.cumsum()
timelines["kill_inner_100_sum"] = timelines.groupby(level=0).kill_inner_100.cumsum()
timelines["kill_inner_200_sum"] = timelines.groupby(level=0).kill_inner_200.cumsum()
```

## Create training and test sets

```python
## randomly draw training and test data
np.random.seed(42)
sample_size = int(0.5 * len(timelines.index.unique()))
train_index = np.random.choice(timelines.index.unique(), size=sample_size, replace=False)
test_index = np.setdiff1d(timelines.index.unique(), train_index)
```

```python

## TODO: check why there are two missing entries for dragons_100/200/diff
## remove this line
timelines = timelines.dropna(axis = 0)
```

```python
X = timelines[["gold_diff",  "gold_100_rel","gold_100", "gold_200", "gold_200_rel", "gold_diff_rel", 
               "dragons_100", "dragons_200",  "dragons_diff", 
               "level_max_100", "level_max_200", "level_sum_100", "level_sum_200", "level_sum_diff", "level_max_diff", 
               "champion_kill_100", "champion_kill_200", "champion_kill_100_sum", "champion_kill_200_sum", "champion_kill_sum_diff", 
               "building_kill_100", "building_kill_200", "kill_inhibitor_100", "kill_inhibitor_200", "building_kill_200", "building_kill_100",
               "inhibitor_kill_100_sum", "inhibitor_kill_200_sum", "kill_inner_100", "kill_inner_200", "kill_inner_100_sum", "kill_inner_200_sum"
               ]]
# X = timelines[["gold_diff", "level_sum_diff", "champion_kill_sum_diff", "dragons_diff"]]
#X = timelines[["gold_diff_rel"]]
X_train = X.ix[train_index]
X_test = X.ix[test_index]
y = timelines["winner_100"]
y_train = y.ix[train_index]
y_test = y.ix[test_index]
```

## Fit random forest

```python
X_train.shape
```

```python
rf = RandomForestClassifier(n_estimators=200, verbose=0, min_samples_leaf=100, 
                            min_samples_split=100, random_state=42, n_jobs=2)
rf.fit(X_train, y_train)
```

## Evaluate forest

```python
## Feature importance
importances = zip(rf.feature_importances_, X.columns)
importances.sort(key=lambda tup: tup[0])
importances
```

## Plot single match

```python
plot_data = timelines.ix[train_index]
plot_data["prediction"] =  [x[1] for x in rf.predict_proba(X_train)]
```

```python
matchId = plot_data.index.unique()[130]
match = plot_data.ix[matchId]
match.plot(x="timestamp", y="prediction", ylim=(0,1))
```
