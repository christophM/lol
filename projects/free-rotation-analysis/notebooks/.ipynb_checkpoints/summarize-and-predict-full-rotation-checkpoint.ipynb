# Summarise rotation data and make predictions

As input I used summmaries for champs and single predictions if a champs will be in the next rotation. Based on that I tried to find a good combination of champions to predict a full rotation. Unfortunately this approach did not work too well.

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import euclidean_distances
import matplotlib
```

```python
rotation_times = pd.read_csv("../data/rotation_times_s4_s5.csv")
```

```python
rotations = rotation_times[rotation_times["event"] == 1]
```

```python
def summarize_rotation(champs):
    """
    Given ten champions, summarize the rotation
    """
    tank = champs[champs.main_role == "Tank"]
    fighter = champs[champs.main_role == "Fighter"]
    marksman= champs[champs.main_role == "Marksman"]
    support = champs[champs.main_role == "Support"]
    mage = champs[champs.main_role == "Mage"]
    assassin = champs[champs.main_role == "Assassin"]

    res = pd.Series({
        "mean_difficulty": np.mean(champs.difficulty),
        "n_tanks": len(tank), 
        "n_fighter": len(fighter),
        "n_marksmen": len(marksman),
        "n_support": len(support),
        "n_mages": len(mage),
        "n_assasin": len(assassin),
        "mean_proba": np.mean(champs.prob)
    })
    return res

class RotationSimilarity():
    """
    Class to find rotations similiar to the stereotypical rotations
    """
    def fit(self, training):
        self.stereotype = summarize_rotation(training)
        
    def predict(self, summarized_rotations):
        return euclidean_distances(self.stereotype, summarized_rotations)[0]
    
    def transform(self):
        pass
    
    
def draw_rotation(champs, selection_id):
    """
    Draws a rotation at random, give a dataframe of champions
    """
    weights = champs.prob.values
    weights = weights / sum(weights)
    ids = np.random.choice(champs.index, 10, replace=False, p=weights)
    champs_selection = champs.ix[ids]
    champs_selection["selection_id"] = selection_id
    return champs_selection

def draw_rotations(champs, n=10000):
    """ Draw n rotations """
    return pd.concat([draw_rotation(champs, i) for i in range(0, n)])
```

```python
## Build stereotype of rotation
summarized_rotations = rotations.groupby("date").apply(summarize_rotation)
stereotype = summarized_rotations.apply(np.mean)
```

```python
summarized_rotations["similarity"] = cosine_similarity(stereotype, summarized_rotations)[0]
```

## Create artificial rotations

```python
rs = RotationSimilarity()
rs.fit(rotations)
## Test for the last rotation
last_date = rotation_times.date.max()
one_rotation = rotation_times[rotation_times.date == last_date]
simulated_rotations = draw_rotations(one_rotation)
percent_correct = simulated_rotations.groupby("selection_id").event.agg(np.mean)
```

```python
summarized_simulated_rotations = simulated_rotations.groupby("selection_id").apply(summarize_rotation)
summarized_simulated_rotations["similarity"] = rs.predict(summarized_simulated_rotations)
summarized_simulated_rotations = summarized_simulated_rotations.join(percent_correct)
```

```python
np.corrcoef(summarized_simulated_rotations.event, summarized_simulated_rotations.mean_proba)
```
