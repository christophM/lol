```python
import pandas as pd
import numpy as np
import lifelines
import matplotlib.pyplot as plt
%matplotlib inline
```

Read raw rotations data

```python
rotations_raw = pd.read_csv("/Users/chris/LOL/lol-free-rotation-analysis/data/historic-raw-rotations.csv", 
                        parse_dates =["date"])
```

```python
champs_static = pd.read_csv("/Users/chris/LOL/lol-free-rotation-analysis/data/champs-static-data.csv", index_col="Champion")
```

DataFrame with information which champions where part of which rotation

```python
rotations_occurence = rotations_raw.copy()
rotations_occurence["event"] = 1
rotations_occurence.drop("season", axis = 1, inplace = True)
rotations_occurence.set_index(["date", "Champion"], inplace = True)
```

Season data

```python
seasons = rotations_raw.reset_index().groupby("date").aggregate(lambda x: x["season"].unique())
seasons.drop(["Champion", "index"], axis = 1, inplace = True)
```

Build a grid of champions and rotation dates

```python
rotations_dates = rotations_raw.date.unique()
champions = rotations_raw.Champion.unique()
champ_date_matrix = pd.DataFrame(np.transpose([np.tile(rotations_dates, len(champions)), np.repeat(champions, len(rotations_dates))]))
champ_date_matrix.columns = ["date", "Champion"]
```

Merge rotation data to grid

```python
rotations_raw["date2"] = rotations_raw.date
## Create date champ matrix
all_rots = pd.merge(champ_date_matrix, rotations_raw, how = "left")
```

Build date x champ - matrix with weeks since last free

```python
date_champ_pivot = all_rots.pivot("date", "Champion", "date2")
date_champ_pivot_shifted = date_champ_pivot.shift(1)
date_champ_pivot_shifted.fillna(method = "pad", inplace = True)
def weeks_since_last_free(champion_dates):
    return (champion_dates.index.values - champion_dates.values) / np.timedelta64(1, 'W')
date_champ_pivot_shifted = date_champ_pivot_shifted.apply(weeks_since_last_free)
date_champ_pivot_shifted.reset_index(inplace = True)
time_since_df = pd.melt(date_champ_pivot_shifted, id_vars = ["date"], value_name = "weeks")
time_since_df.set_index(["date", "Champion"], inplace = True)
```

Merge with information if champ was really in rotation

```python
time_since_df = time_since_df.join(rotations_occurence)
time_since_df["event"].fillna(0, inplace = True)
```

```python

## TODO: 
# Do not throw away all occurences here. keep those with weeks == NaN but event == 1

time_since_df.weeks.fillna(-1, inplace = True)
```

```python
rotation_times = time_since_df.join(seasons)
```

```python
rotation_times = rotation_times.join(champs_static)
rotation_times = rotation_times.reset_index()
```

```python
ranks = rotation_times.groupby(["date", "main_role"]).weeks.rank(ascending=False)
ranks.name = "wait_rank_by_role"
rotation_times = rotation_times.join(ranks)
```

```python
freqs = rotation_times.sort("date").groupby("Champion").event.apply(lambda x: np.cumsum(x) )
freqs.name = "freqs"
rotation_times = rotation_times.join(freqs)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_recall_fscore_support
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
np.random.seed(42)
train_index = np.random.choice(rotation_times.index, 0.5 * len(rotation_times))
train_index1 = np.random.choice(train_index, 0.5 * len(train_index))
train_index2 = np.setdiff1d(train_index, train_index1)
test_index = np.setdiff1d(rotation_times.index.values, train_index)
```

```python
class FeatureBuilder():
    def fit(self):
        pass
    def transform(self, data):
        input_columns = ["weeks",  "main_role", "freqs", "difficulty", "attack", "defense", "magic", "wait_rank_by_role"]
        X = data[input_columns]
        X = X.join(pd.core.reshape.get_dummies(X.main_role, prefix="main"))
        X.drop([ "main_role"], axis=1, inplace=True)
        return X
        
```

```python
fb = FeatureBuilder()
```

```python
X = fb.transform(rotation_times)
y = rotation_times.event
```

### Fit random forest

```python
mod = RandomForestClassifier()
mod.fit(X.ix[train_index1], y.ix[train_index1])
```

```python
zip(mod.feature_importances_, X.columns)
```

## Evaluate goodness

```python
y_true = y.ix[train_index2]
y_pred = mod.predict(X.ix[train_index2])
precision_recall_fscore_support(y_true, y_pred)
from sklearn.metrics import roc_auc_score
y_proba = [x[1] for x in mod.predict_proba(X.ix[train_index2])]
roc_auc_score(y_true, y_proba)
```

```python
last_date = rotation_times.date.max()
```

```python
remove_index = rotation_times.index[rotation_times.date == last_date]
my_index = np.setdiff1d(X.index.values, remove_index.values)
```

```python
mod.fit(X.ix[my_index], y.ix[my_index])
rotation_times["prob"] = [x[1] for x in mod.predict_proba(X)]
```
