# Download additional champion information via API

This script downloads static champion information using the rawpi api.  
The champ infos include the description of spells, the strengths and so on. 

This script is used to download the data. It is not stored anywhere. I tried to fit a Topic model (LDA) to identify similar spells which did not work out, because similar spells are described with diffrent words. 

```python
import rawpi
import json
import pandas as pd
import numpy as np
from gensim import corpora, models, similarities
import nltk
from nltk.corpus import stopwords
from operator import itemgetter
```

## Load champion static data

```python
champs_raw = rawpi.get_champion_list(region="euw", champData="all").text
champs = json.loads(champs_raw)["data"]
```

```python
def cleanup_description(description, champ):
    """
    Clean the description field of a champ
    Removes numbers and special characters
    INPUT: description field, name of champ
    OUTPUT: cleaned description field
    """
    descr = description.lower()
    ## remove placeholder 
    descr = re.sub(r'\{\{ [a-z][0-9] \}\}', '', descr)
    ## remove punctuation
    descr = re.sub(r",|\(|\)", "", descr)
    ## replace numbers with dummy
    descr = re.sub(r"[0-9]+\.?[0-9]", "NUM", descr)
    descr = re.sub("mundo|wukong|%s" % champ.lower(), 'thischamp', descr)
    return descr
```

## Parse abilities of champions

```python
abilities_df = pd.DataFrame()
for champ, attributes in champs.iteritems():
    spells = attributes["spells"]
    descriptions = [ cleanup_description(spell["sanitizedTooltip"], champ) for spell in spells]
    names = [spell["name"] for spell in spells]
    abilities = pd.DataFrame({
        "description": descriptions, 
        "name": names,
        "champ": np.repeat(champ, len(spells))
    }, index=[spell["key"] for spell in spells])
    abilities_df = abilities_df.append(abilities)
```

## Try LDA

```python
documents = abilities_df.description
stoplist = stopwords.words('english')

def ngrams(text, n=1):
  "Input format: list of strings"
  output = []
  for i in range(len(text)-n+1):
    output.append(" ".join(text[i:i+n]))
  return output

def sentence_to_ngrams(sentence):
    words = [word for word in sentence.split() if word not in stoplist]
    return ngrams(words)

def doc_to_ngrams(doc):
    sentences = doc.lower().split(".")
    return reduce(lambda x,y: x + sentence_to_ngrams(y), sentences, [])
    

texts = [doc_to_ngrams(document) for document in documents]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

tfidf = gensim.models.TfidfModel(corpus) 
corpus_tfidf = tfidf[corpus]

#lsi = models.LsiModel(corpus_tfidf, id2word=dictionary, num_topics=100)
#lsi.print_topics(20)

n_topics = 20
lda = models.LdaModel(corpus_tfidf, id2word=dictionary, num_topics=n_topics, iterations=1000)
```

```python
for i in range(0, n_topics):
    temp = lda.show_topic(i, 10)
    terms = [term[1] for term in temp]
    print "\nTop 10 terms for topic #" + str(i) + ":\n" + ", \n".join(terms)

print 'Which LDA topic maximally describes a document?\n'
print 'Original document: ' + documents[1]
print 'Preprocessed document: ' + str(texts[1])
print 'Topic probability mixture: ' + str(lda[corpus[1]])
print 'Maximally probable topic: topic #' + str(max(lda[corpus[1]],key=itemgetter(1))[0])
```
